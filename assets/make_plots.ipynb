```python
import numpy as np

import matplotlib.pyplot as plt

# Define the age range
ages = np.arange(18, 44)

# Calculate the potential loss
monthly_payment = 100
potential_loss = (43 - ages) * 12 * monthly_payment

# Plot the data
plt.figure(figsize=(10, 6), facecolor='none')
plt.plot(ages, potential_loss, color='#8ccd10')
plt.title('Potential Loss in Insurance Payments by Age', color='white')
plt.xlabel('Age', color='white')
plt.ylabel('Potential Loss ($)', color='white')
plt.grid(True, color='gray')
plt.gca().patch.set_alpha(0)  # Make the background transparent
plt.gca().spines['bottom'].set_color('white')
plt.gca().spines['top'].set_color('white')
plt.gca().spines['right'].set_color('white')
plt.gca().spines['left'].set_color('white')
plt.gca().tick_params(axis='x', colors='white')
plt.gca().tick_params(axis='y', colors='white')
plt.show()
```
