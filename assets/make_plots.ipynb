```python
import numpy as np

import matplotlib.pyplot as plt

# Define the age range
ages = np.arange(18, 44)

# Calculate the potential loss
monthly_payment = 100
potential_loss = (43 - ages) * 12 * monthly_payment

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(ages, potential_loss, marker='o')
plt.title('Potential Loss in Insurance Payments by Age')
plt.xlabel('Age')
plt.ylabel('Potential Loss ($)')
plt.grid(True)
plt.show()
```
